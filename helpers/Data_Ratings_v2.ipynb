```python
def create_combinations(row, columns):
    list_of_entries = []
    for col in columns:
        entries = str(row[col]).split() if pd.notna(row[col]) else []
        list_of_entries.append(entries if entries else [None])
    
    all_combinations = list(product(*list_of_entries))
    
    row_data = row.to_dict()
    expanded_rows = []
    for combo in all_combinations:
        new_row = row_data.copy()
        for i, col in enumerate(columns):
            new_row[col] = combo[i]
        expanded_rows.append(new_row)
        
    return expanded_rows

columns_to_combine = [
    'actor', 'actress', 'archive_footage', 'archive_sound',
    'cinematographer', 'composer', 'director', 'editor',
    'producer', 'production_designer', 'self', 'writer'
]

expanded_data_all_cols = []
for index, row in merged_df_rev.iterrows():
    expanded_data_all_cols.extend(create_combinations(row, columns_to_combine))

expanded_merged_df_rev_all_cols = pd.DataFrame(expanded_data_all_cols)

print(expanded_merged_df_rev_all_cols)
```

```python
#split by genre


def expand_genre(row):
    genres = row['genres'].split(',') if pd.notna(row['genres']) else [None]

    row_data = row.to_dict()
    expanded_rows = []
    for genre in genres:
        new_row = row_data.copy()
        new_row['genres'] = genre
        expanded_rows.append(new_row)
        
    return expanded_rows

expanded_data_genres = []
for index, row in expanded_merged_df_rev_all_cols.iterrows():
    expanded_data_genres.extend(expand_genre(row))

df_clean = pd.DataFrame(expanded_data_genres)

print(df_clean)
```
