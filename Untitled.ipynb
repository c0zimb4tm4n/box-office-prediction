```python
import pandas as pd
```

```python
principals=pd.read_csv('Data/title.principals.tsv',sep='\t')
basics=pd.read_csv('Data/title.basics.tsv',sep='\t')
ratings=pd.read_csv('Data/title.basics.tsv',sep='\t')
```

```python
actors = principals[principals["category"].str.contains("actor|self")]
```

```python
#actors_pivot = actors.pivot(columns = ["tconst","ordering"], values = "nconst")
actors_pivot = actors.pivot(index='tconst', columns='ordering', values='nconst').reset_index()

actors_pivot.columns = [f'nconst_{col}' if isinstance(col, int) else col for col in actors_pivot.columns]
```

```python
actors_pivot.head(100)
```

```python
actors_pivot.columns.values
merged = basics.merge(actors_pivot, on="tconst")
merged.columns
```

```python
all_types = merged[['tconst', 'titleType', 'primaryTitle', 'originalTitle', 'isAdult',
       'startYear', 'endYear', 'runtimeMinutes', 'genres', 'nconst_1',
       'nconst_2', 'nconst_3', 'nconst_4', 'nconst_5']]
all_types.head()
```
